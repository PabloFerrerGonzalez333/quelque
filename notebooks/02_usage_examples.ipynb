# Quelque Tutorial 2: How To Use The Project

This notebook is a practical tutorial. It explains how to install, run, test, and troubleshoot the repository.

It is written as onboarding material for someone who just cloned the repo.


## 1. Choose The Right Workflow First

Before running anything, decide what kind of experience you want:

### Best demo quality

Use hosted mode.

- better transcription quality
- better summaries
- better fit for portfolio screenshots and demos

### Best offline / no-key path

Use local mode.

- no OpenAI key required
- slower on CPU
- output quality depends more heavily on the audio and local model


## 2. Local Setup

On Windows PowerShell:

```powershell
python -m venv .venv
.venv\Scripts\Activate.ps1
pip install -e .[dev,docs]
Copy-Item .env.example .env
```

If you want hosted mode, set `OPENAI_API_KEY` inside `.env`.


## 3. Run The Streamlit App

```powershell
streamlit run streamlit_app.py
```

What to do in the UI:

1. Upload a lecture recording.
2. Choose hosted or local mode.
3. Select a language hint if you know it.
4. Generate notes.
5. Review the transcript and export results.

For a portfolio demo, hosted mode is the strongest default.


## 4. CLI Usage

### Local transcription on the real class sample

```powershell
quelque transcribe samples/class_cropped.mp3 --provider local --language fr
```

### Local notes on the same sample

```powershell
quelque notes samples/class_cropped.mp3 --transcription-provider local --notes-provider local
```

### Hosted mode

```powershell
quelque notes samples/class_cropped.mp3 --transcription-provider openai --notes-provider openai --language fr
```


## 5. What To Expect From Local Mode

This is important and should be explicit:

- the first local run can take a long time because the speech model may download
- CPU transcription on lecture audio can take several minutes
- noisy academic audio often produces rougher transcripts locally
- the local notes provider is intentionally simple and mainly exists as a fallback

So if local output looks rough, that does **not** necessarily mean the app is broken. It often means the fallback path is behaving as designed.


## 6. Python API Example

If you want to use the repository programmatically instead of through the UI or CLI, use the pipeline directly.


````python
from pathlib import Path

from quelque.config import AppConfig
from quelque.pipeline import QuelquePipeline

config = AppConfig()
pipeline = QuelquePipeline(config)

result = pipeline.run(
    Path('samples/class_cropped.mp3'),
    transcription_provider='local',
    notes_provider='local',
    language_hint='fr',
)

print('Cache hit:', result.cache_hit)
print('Elapsed:', round(result.elapsed_seconds, 2), 'seconds')
print('Detected language:', result.notes.transcript.detected_language)
print('Summary preview:', result.notes.summary[:300])
````

## 7. Troubleshooting

### Problem: local commands seem stuck

Possible causes:

- first-time model download
- CPU transcription on a multi-minute file
- large audio input

### Problem: transcript quality is weak

Try:

- adding a language hint such as `--language fr`
- using hosted mode instead of local mode
- testing on cleaner audio

### Problem: the tiny sample produces empty notes

That is expected, because the tiny sample is synthetic and mainly meant for smoke tests, not for a realistic demo.


## 8. Recommended Portfolio Demo Flow

If you want to show this project to someone else, the strongest sequence is:

1. explain the problem in one sentence
2. open the Streamlit app
3. upload a short real lecture sample
4. show the transcript tab
5. show the notes tab
6. export Markdown or PDF
7. explain why hosted mode is the product path and local mode is the fallback path
