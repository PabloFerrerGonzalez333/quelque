## 1. The Problem Quelque Solves

Students and researchers often end up with long recordings: lectures, seminars, study sessions, oral explanations, and class captures.

Those recordings are useful, but they are hard to search, hard to review, and slow to turn into study materials.

Quelque exists to convert that raw audio into something more usable:

- a transcript
- a concise summary
- key takeaways
- study notes
- glossary items
- action items


## 2. The Main Product Idea

At a high level, the app follows this sequence:

1. Load an audio file.
2. Normalize it to a format suitable for speech recognition.
3. Split long recordings into chunks.
4. Transcribe the audio.
5. Turn the transcript into study-friendly notes.
6. Export the result as Markdown, DOCX, or PDF.

This makes the repository useful both as a real tool and as a portfolio project that demonstrates productization work, not just experimentation.

## 4. Hosted Mode vs Local Mode

This is one of the most important concepts in the repository.

### Hosted mode

Hosted mode is the recommended path for demos and portfolio presentation.

Why:

- it is faster
- it usually gives better transcription quality
- it better represents the intended product experience

Tradeoff:

- it requires API credentials

### Local mode

Local mode is the fallback path.

Why it exists:

- offline usage
- no API key required
- useful for development and smoke testing

Tradeoffs:

- slower on CPU
- first run downloads the model
- transcription quality may be rougher, especially on noisy or academic lecture audio


## 5. What We Learned From The Sample Audio

The repository currently includes two useful example files:

- `samples/tiny_sanitized_sample.wav`: a tiny synthetic fixture for tests
- `samples/class_cropped.mp3`: a real speech sample that is better for demonstration

Important distinction:

- the tiny fixture is good for automated testing
- the class sample is better for human-facing demos

This matters because a tone-only fixture can make the pipeline look broken even when it is actually behaving correctly.
